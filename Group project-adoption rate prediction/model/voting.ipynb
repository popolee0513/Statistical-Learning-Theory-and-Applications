```python
import pandas as pd
import random
import math
import os
```

```python
path=os.path.abspath('..')
```

```python
vote1=pd.read_csv(path+"\\result\\gp_and_bagging.csv")
vote2=pd.read_csv(path+"\\result\\gp_and_bagging_without_description.csv")
vote3=pd.read_csv(path+"\\result\\logistic_des.csv")
vote4=pd.read_csv(path+"\\result\\logistic_mcmc.csv")
vote5=pd.read_csv(path+"\\result\\logistic_nodes.csv")
vote6=pd.read_csv(path+"\\result\\RF_prediction.csv", header=None)
vote7=pd.read_csv(path+"\\result\\xgboost.csv", header=None)
vote8=pd.read_csv(path+"\\result\\xg-without-des.csv", header=None)
vote9=pd.read_csv(path+"\\result\\NB_with_des.csv", header=None)
vote10=pd.read_csv(path+"\\result\\NB_without_des.csv", header=None)
vote11=pd.read_csv(path+"\\result\\logistic_with_des.csv", header=None)
vote12=pd.read_csv(path+"\\result\\logistic_without_des.csv", header=None)
```

```python
vote1 = vote1.rename(index=str, columns={"gp": "vote1a", "bag": "vote1b"})
vote2 = vote2.rename(index=str, columns={"gp": "vote2a", "bag": "vote2b"})
vote3 = vote3.rename(index=str, columns={"0": "vote3"})
vote4 = vote4.rename(index=str, columns={"0": "vote4"})
vote5 = vote5.rename(index=str, columns={"0": "vote5"})
vote6 = vote6.rename(index=str, columns={"0": "vote6"})
vote7 = vote7.rename(index=str, columns={"0": "vote7"})
vote8 = vote8.rename(index=str, columns={"0": "vote8"})
vote9 = vote9.rename(index=str, columns={"0": "vote9"})
vote10 = vote10.rename(index=str, columns={"0": "vote10"})
vote11 = vote11.rename(index=str, columns={"0": "vote11"})
vote12 = vote12.rename(index=str, columns={"0": "vote12"})
```

```python
vote3["vote3"] = vote3["vote3"].map( lambda x: x-1 )
vote4["vote4"] = vote4["vote4"].map( lambda x: x-1 )
vote5["vote5"] = vote5["vote5"].map( lambda x: x-1 )
```

```python
result = pd.concat([vote1, vote2, vote3, vote4, vote5, vote6, vote7, vote8, vote9, vote10, vote11, vote12], axis=1, sort=False)
```

```python
predict = result.mode(axis='columns')
```

```python
predict
```

```python
index = 0
while index < predict.shape[0]:
    temp = predict[ math.floor(random.random()*predict.count(axis='columns')[0]) ][index]
    predict[0][index] = temp
    index+=1
```

```python
from sklearn.model_selection import train_test_split
data=pd.read_csv(path+"\\data\\train.csv")
X = data
y = data['AdoptionSpeed']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
ans = pd.DataFrame([]) 
ans = pd.concat([ans, y_test], axis=1, ignore_index = True )
ans = ans.reset_index()[0]
```

```python
def accuracy(pred,ans):
    count=0
    for i in range(len(pred)):
        if pred[i]==ans[i]:
            count+=1
    return count/len(pred)
accuracy(predict[0],ans)
```
